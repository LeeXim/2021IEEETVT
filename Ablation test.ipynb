```python
import math
v_reliability = []

v_std = [np.std(V_list[0:9]),np.std(V_list[9:18])
          ,np.std(V_list[18:26]),np.std(V_list[26:30]),np.std(V_list[30:34]), np.std(V_list[34:36])]

for i in v_std:
    v_reliability.append(1. * i / math.sqrt(6))
    
v21_reliability = []

v21_std = [np.std(V21_list[0:8]),np.std(V21_list[8:16])
          ,np.std(V21_list[16:24]),np.std(V21_list[24:28]),np.std(V21_list[28:32]), np.std(V21_list[32:34])]

for i in v21_std:
    v21_reliability.append(1. * i / math.sqrt(6))
    
v24_reliability = []

v24_std = [np.std(V24_list[0:8]),np.std(V24_list[8:16])
          ,np.std(V24_list[16:24]),np.std(V24_list[24:28]),np.std(V24_list[28:32]), np.std(V24_list[32:34])]

for i in v24_std:
    v24_reliability.append(1. * i / math.sqrt(6))
```

```python
upper_ci = []
lower_ci = []

v_mean = [np.mean(V_list[0:9]),np.mean(V_list[9:18])
          ,np.mean(V_list[18:26]),np.mean(V_list[26:30]),np.mean(V_list[30:34]),np.mean(V_list[34:36])]

for i in range(len(v_mean)):
    upper_ci.append(v_mean[i] + v_reliability[i])
    lower_ci.append(v_mean[i] - v_reliability[i])
    
upper21_ci = []
lower21_ci = []

v21_mean = [np.mean(V21_list[0:8]),np.mean(V21_list[8:16])
          ,np.mean(V21_list[16:24]),np.mean(V21_list[24:28]),np.mean(V21_list[28:32]),np.mean(V21_list[32:34])]

for i in range(len(v21_mean)):
    upper21_ci.append(v21_mean[i] + v21_reliability[i])
    lower21_ci.append(v21_mean[i] - v21_reliability[i])
    
upper24_ci = []
lower24_ci = []

v24_mean = [np.mean(V24_list[0:8]),np.mean(V24_list[8:16])
          ,np.mean(V24_list[16:24]),np.mean(V24_list[24:28]),np.mean(V24_list[28:32]),np.mean(V24_list[32:34])]

for i in range(len(v24_mean)):
    upper24_ci.append(v24_mean[i] + v24_reliability[i])
    lower24_ci.append(v24_mean[i] - v24_reliability[i])
```
